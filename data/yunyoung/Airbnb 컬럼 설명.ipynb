### listing 컬럼 설명
```
id: 각 목록에 Airbnb에서 할당한 고유 식별자입니다.
listing_url: Airbnb 웹사이트에서 해당 목록의 URL입니다.
scrape_id: Inside Airbnb에서 수행한 특정 스크래이핑 작업을 식별하기 위한 식별자입니다.
last_scraped: 목록이 마지막으로 스크래이핑된 날짜와 시간입니다.
source: 목록이 동네 검색을 통해 발견되었는지 아니면 이전 스크래이핑에서 확인된 목록인지를 나타냅니다. name: 목록의 이름/제목입니다.
description: 목록에 대한 상세한 설명입니다.
neighborhood_overview: 호스트가 작성한 동네에 대한 설명입니다.
picture_url: 목록을 위해 Airbnb에서 호스팅하는 정규 크기 이미지의 URL입니다.
host_id: 호스트/사용자에게 Airbnb에서 할당된 고유 식별자입니다.
 host_url: 호스트의 Airbnb 페이지입니다.
 host_name: 호스트의 이름입니다(일반적으로 이름(들)으로 표시됩니다).
 host_since: 호스트/사용자가 Airbnb에 가입한 날짜입니다. 호스트가 Airbnb 게스트인 경우, 게스트로 등록한 날짜일 수도 있습니다.
host_location: 호스트가 자체적으로 보고한 위치입니다.
host_about: 호스트에 대한 설명입니다.
host_response_time: 호스트의 문의 응답 시간의 평균입니다.
host_response_rate: 호스트의 문의 응답률입니다.
host_acceptance_rate: 호스트가 예약 요청을 수락하는 비율입니다. host_is_superhost: 호스트가 슈퍼호스트 상태에 도달했는지 여부를 나타내는 불리언 값입니다(참/거짓).
host_thumbnail_url: 호스트의 썸네일 크기 이미지의 URL입니다.
host_picture_url: 호스트의 프로필 사진의 URL입니다.
host_neighbourhood: 호스트가 위치한 동네입니다.
host_listings_count: Airbnb의 계산에 따라 호스트가 가지고 있는 목록의 수입니다. host_total_listings_count: 호스트가 가지고 있는 전체 목록의 수입니다. host_verifications: 호스트가 사용하는 인증 방법입니다.
host_has_profile_pic: 호스트가 프로필 사진을 가지고 있는지 여부를 나타내는 불리언 값입니다(참/거짓).
host_identity_verified: 호스트의 신원이 확인되었는지 여부를 나타내는 불리언 값입니다(참/거짓).
neighbourhood: 목록이 위치한 동네입니다.
neighbourhood_cleansed: 위도와 경도를 사용하여 지오코드된 동네입니다. neighbourhood_group_cleansed: 위도와 경도를 사용하여 지오코드된 동네 그룹입니다.
latitude: 목록의 위치에 대한 위도 좌표입니다.
longitude: 목록의 위치에 대한 경도 좌표입니다.
property_type: 호스트가 선택한 자체 속성 유형입니다.
room_type: 예약 가능한 방의 유형입니다(예: Entire home/apt, Private room, Shared room, Hotel).
accommodates: 목록의 최대 수용 인원입니다.
bathrooms: 목록에 있는 욕실의 수입니다.
bathrooms_text: 목록에 있는 욕실의 수에 대한 텍스트 설명입니다.
bedrooms: 목록에 있는 침실의 수입니다.
beds: 목록에 있는 침대의 수입니다. amenities: 제공되는 편의 시설 목록을 포함한 JSON 객체입니다.
price: 목록의 일일 가격(현지 통화)입니다.
minimum_nights: 예약을 위해 필요한 최소 숙박일 수입니다.
maximum_nights: 예약이 허용되는 최대 숙박일 수입니다.
minimum_minimum_nights: 캘린더에서 가장 작은 최소 숙박일 값입니다(미래 365일 동안 확인).
maximum_minimum_nights: 캘린더에서 가장 큰 최소 숙박일 값입니다(미래 365일 동안 확인).
minimum_maximum_nights: 캘린더에서 가장 작은 최대 숙박일 값입니다(미래 365일 동안 확인).
maximum_maximum_nights: 캘린더에서 가장 큰 최대 숙박일 값입니다(미래 365일 동안 확인).
minimum_nights_avg_ntm: 캘린더에서 평균 최소 숙박일 값입니다(미래 365일 동안 확인).
maximum_nights_avg_ntm: 캘린더에서 평균 최대 숙박일 값입니다(미래 365일 동안 확인).
calendar_updated: 캘린더가 마지막으로 업데이트된 날짜입니다.
has_availability: 목록의 가용성 여부를 나타내는 불리언 값입니다(참/거짓). availability_30: 목록의 다음 30일간 가용성입니다.
availability_60: 목록의 다음 60일간 가용성입니다.
availability_90: 목록의 다음 90일간 가용성입니다.
availability_365: 목록의 다음 365일간 가용성입니다.
calendar_last_scraped: 캘린더가 마지막으로 스크래이핑된 날짜입니다. number_of_reviews: 목록이 받은 총 리뷰 수입니다.
number_of_reviews_ltm: 목록이 받은 최근 12개월간 리뷰 수입니다. number_of_reviews_l30d: 목록이 받은 최근 30일간 리뷰 수입니다.
first_review: 목록의 첫 번째/가장 오래된 리뷰의 날짜입니다.
 last_review: 목록의 마지막/최신 리뷰의 날짜입니다.
review_scores_rating: 게스트 리뷰에 기반한 목록의 전체 평점 점수입니다. review_scores_accuracy: 게스트 리뷰에 기반한 목록의 정확성 평점 점수입니다. review_scores_cleanliness: 게스트 리뷰에 기반한 목록의 청결도 평점 점수입니다. review_scores_checkin: 게스트 리뷰에 기반한 목록의 체크인 절차 평점 점수입니다. review_scores_communication: 게스트 리뷰에 기반한 목록의 의사소통 평점 점수입니다.
review_scores_location: 게스트 리뷰에 기반한 목록의 위치 평점 점수입니다. review_scores_value: 게스트 리뷰에 기반한 목록의 가치 평점 점수입니다.
license: 목록의 라이선스, 허가 또는 등록 번호입니다.
instant_bookable: 목록을 호스트의 승인 없이 즉시 예약할 수 있는지 여부를 나타내는 불리언 값입니다(참/거짓).
calculated_host_listings_count: 현재 스크래이핑에서 호스트가 가지고 있는 목록의 수(도시/지역 기준)입니다.
calculated_host_listings_count_entire_homes: 현재 스크래이핑에서 호스트가 가지고 있는 전체 주택/아파트 목록의 수(도시/지역 기준)입니다. calculated_host_listings_count_private_rooms: 현재 스크래이핑에서 호스트가 가지고 있는 개인실 목록의 수(도시/지역 기준)입니다. calculated_host_listings_count_shared_rooms: 현재 스크래이핑에서 호스트가 가지고 있는 공용실 목록의 수(도시/지역 기준)입니다.
reviews_per_month: 목록이 수신한 리뷰의 평균 월간 수입니다.

```
