```python
import glob

data = glob.glob('tokyo/*')
```

```python
import pandas as pd

tokyoSiryo = pd.DataFrame()
for idx in range(len(data)):
    if 'siryo' in data[idx]:
        df = pd.read_csv(data[idx], encoding='SHIFT_JIS')
        tokyoSiryo  = pd.concat([tokyoSiryo ,df])
tokyoSiryo = tokyoSiryo .drop(columns = 'Unnamed: 0')
```

```python
tokyoSiryo .reset_index(drop=True, inplace=True)
```
