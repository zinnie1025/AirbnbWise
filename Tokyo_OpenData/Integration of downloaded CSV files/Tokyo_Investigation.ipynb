```python
import glob

data = glob.glob('tokyo/*')
```

```python
import pandas as pd

Investigation = pd.DataFrame()
for idx in range(len(data)):
    if 'investigation' in data[idx]:
        df = pd.read_csv(data[idx], encoding='SHIFT_JIS')
        Investigation  = pd.concat([Investigation ,df])
Investigation = Investigation .drop(columns = 'Unnamed: 0')
```

```python
Investigation.reset_index(drop=True, inplace=True)
```
