```python
import glob

data = glob.glob('tokyo/*')
```

```python
import pandas as pd

tokyoDocument = pd.DataFrame()
for idx in range(len(data)):
    if '02_document' in data[idx]:
        df = pd.read_csv(data[idx], encoding='SHIFT_JIS')
        tokyoDocument = pd.concat([tokyoDocument,df])
tokyoDocument = tokyoDocument.drop(columns = 'Unnamed: 0')
```

```python
tokyoDocument.reset_index(drop=True, inplace=True)
```
