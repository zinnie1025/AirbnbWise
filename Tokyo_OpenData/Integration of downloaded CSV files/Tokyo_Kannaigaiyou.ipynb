```python
import glob

data = glob.glob('tokyo/*')
```

```python
import pandas as pd
from tqdm import tqdm_notebook

tokyoKannaigaiyou = pd.DataFrame()
for idx in range(len(data)):
    if 'kannaigaiyou' in data[idx]:
        df = pd.read_csv(data[idx], encoding='SHIFT_JIS')
        tokyoKannaigaiyou   = pd.concat([tokyoKannaigaiyou  ,df])
tokyoKannaigaiyou = tokyoKannaigaiyou .drop(columns = 'Unnamed: 0')
```

```python
tokyoKannaigaiyou .reset_index(drop=True, inplace=True)
```
