```python
import glob

data = glob.glob('tokyo/*')
```

```python
import pandas as pd

tokyoReport = pd.DataFrame()
for idx in range(len(data)):
    if 'report' in data[idx]:
        df = pd.read_csv(data[idx], encoding='SHIFT_JIS')
        tokyoReport  = pd.concat([tokyoReport ,df])
tokyoReport = tokyoReport .drop(columns = 'Unnamed: 0')
```

```python
tokyoReport.reset_index(drop=True, inplace=True)
```
