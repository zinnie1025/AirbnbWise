```python
import glob

data = glob.glob('tokyo/*')
```

```python
import pandas as pd

tn_2018 = pd.DataFrame()
tn_2019 = pd.DataFrame()

for idx in range(len(data)):
    if 'tn18' in data[idx]:
        df = pd.read_csv(data[idx], encoding='SHIFT_JIS')
        tn_2018 = pd.concat([tn_2018,df])
    elif 'tn19' in data[idx]:
        df = pd.read_csv(data[idx], encoding='SHIFT_JIS')
        tn_2019 = pd.concat([tn_2019,df])
        
tn_2018 = tn_2018.drop(columns = {'Unnamed: 0', '1'})
tn_2019 = tn_2019.drop(columns = {'Unnamed: 0', '1'})
```

#### 이 데이터는 칼럼명이 명확하지 않아 지칭하는 데이터가 뭔지 알기 어렵고, null값이 대부분이라 사용하지 않는다.

```python
tokyoTn = pd.concat([tn_2018, tn_2019])
tokyoTn.reset_index(drop=True, inplace=True)
tokyoTn
```
