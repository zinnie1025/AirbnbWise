```python
import glob

data = glob.glob('tokyo/*')
```

```python
import pandas as pd

tokyoSupport = pd.DataFrame()
for idx in range(len(data)):
    if 'support' in data[idx]:
        df = pd.read_csv(data[idx], encoding='SHIFT_JIS')
        tokyoSupport = pd.concat([tokyoSupport,df])
tokyoSupport = tokyoSupport.drop(columns = 'Unnamed: 0')
```

```python
tokyoSupport.reset_index(drop=True, inplace=True)
```
