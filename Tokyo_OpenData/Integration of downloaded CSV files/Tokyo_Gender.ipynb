```python
import glob

data = glob.glob('tokyo/*')
```

```python
import pandas as pd

tokyoGender = pd.DataFrame()
for idx in range(len(data)):
    if 'gender-equal' in data[idx]:
        df = pd.read_csv(data[idx], encoding='SHIFT_JIS')
        tokyoGender = pd.concat([tokyoGender,df])
tokyoGender = tokyoGender.drop(columns = 'Unnamed: 0')
```

```python
tokyoGender.reset_index(drop=True, inplace=True)
```

#### 이 데이터는 null값이 대부분이고, 치안과 관련 없는 데이터이므로 사용하지 않는다.
