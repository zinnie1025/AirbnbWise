### 위도 경도 컬럼 생성 
<b>추후 사용자가 선택한 숙소의 반경 내 부동산 사건 사고 정보를 제공하기 위해서는 위도 경도 정보가 필요하다 </b>

<b>haversine 라이브러리를 이용하여  위도 경도 기준 두 위치간 거리를 계산 가능  </b>
 
<b>jageocoder 라이브러리 이용</b> : https://jageocoder.readthedocs.io/ja/latest/index.html

- id: ID in the database
- name: Address notation
- x: longitude
- y: latitude
- level: Address level (1:Prefecture, 2:County, 3:City and 23 district, 4:Ward, 5:Oaza, 6:Aza and - Chome, 7:Block, 8:Building)-
- note: Notes such as city codes
- fullname: List of address notations from the prefecture level to this node

```python
import jageocoder
jageocoder.init()
```

```python
longitude_list = []
latitude_list = []
full_name_list = []

# 위도,경도  
def get_coordinate(address):
    full_name = address
    lat = 0
    lng = 0
    result = jageocoder.search(address)
    candidates = result['candidates'][0]
    full_name = candidates['fullname']
    lng = candidates['x']
    lat = candidates['y']
    longitude_list.append(lng)
    latitude_list.append(lat)
    full_name_list.append(' '.join(full_name))
```

```python
filtered_oshima_data['address'].apply(get_coordinate)
filtered_oshima_data['full name'] = full_name_list
filtered_oshima_data['latitude'] = latitude_list
filtered_oshima_data['longitude'] = longitude_list
filtered_oshima_data.tail()
```

```python
filtered_oshima_data.to_csv('oshima_data_add_latlng.csv')
```

### 입력 데이터에 포함 할 컬럼만 가져옴

```python
input_cols = ['neighbourhood_cleansed', 'price', 'room_type', 'accommodates', 'bathrooms_text', 'bedrooms' , 'beds', 'review_scores_rating', 'latitude', 'longitude']
airbnb_df = listing_gz[input_cols]
airbnb_df.head()
```

#### Address 추가 

```python
def add_jp_address(lng, lat):
    address = jageocoder.reverse(lng, lat)
    if len(address)>0:
        address = ' '.join(address[0]['candidate']['fullname'])
    else:
        address = 'None'

    return address
```

```python
triangle = jageocoder.reverse(139.422913, 35.637493)
if len(triangle) > 0:
    print(' '.join(triangle[0]['candidate']['fullname'])) # 東京都 八王子市 大塚 640番地	
```

```python
address_list = []
for index, row in airbnb_df.iterrows():
    address = add_jp_address(row['longitude'], row['latitude'])
    address_list.append(address)

airbnb_df['full name'] = address_list
airbnb_df.tail()
```

#### Date 형식 

```python
# TODO: 'timeOfCrime' 칼럼 데이터 시간 양식이 동일하지 않다. 양식을 통일 하고, 데이터 타입을 datetime(ns)로 변경한다.
# ex) 2023-07-18 , 2023/07/18
# 추후 year, month, day 칼럼으로 분리하여 추가할 예정이다.
def getDateOfCrime(date):
    if pd.isnull(date):
        return pd.NaT

    if isinstance(date, str):
        if '-' in date:
            dateValue = pd.to_datetime(date, format='%Y-%m-%d', errors='coerce')
        elif '/' in date:
            dateValue = pd.to_datetime(date, format='%Y/%m/%d', errors='coerce')
        else:
            dateValue = pd.to_datetime(date, errors='coerce')
    else:
        dateValue = date

    return dateValue


open_data_filter['発生年月日（始期）'] = open_data_filter['発生年月日（始期）'].apply(getDateOfCrime)
open_data_filter.tail()
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from collections import Counter
import MeCab

m = MeCab.Tagger()


def extract_nouns(text):
    node = m.parseToNode(text)
    nouns = []
    while node:
        if node.feature.split(',')[0] == '名詞':  # Extract only nouns
            nouns.append(node.surface)
        node = node.next
    return nouns

# Apply the function to each row in the 'info' column
oshima_data['nouns'] = oshima_data['info'].apply(extract_nouns)
oshima_data[:10]
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from collections import Counter
import MeCab

# 일본어 폰트 경로 설정 (자신의 환경에 맞게 변경)
font_path = '/System/Library/Fonts/Arial Unicode.ttf'

# MeCab 형태소 분석기 초기화
m = MeCab.Tagger()

# 일본어 문장
japanese_text = "こちらは日本語のテキストです。日本語のワード클라우드를 생성해봅시다。"

# 형태소 분석 후 단어 추출
node = m.parseToNode(japanese_text)
words = []
while node:
    if node.feature.split(',')[0] == '名詞':  # 명사만 추출
        words.append(node.surface)
    node = node.next

# 단어 빈도 카운트
word_counts = Counter(words)

# WordCloud 생성
wordcloud = WordCloud(width=800, height=400, font_path=font_path,
                      background_color='white').generate_from_frequencies(word_counts)

plt.figure(figsize=(12, 8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Japanese Word Cloud\n')
plt.show()
```
